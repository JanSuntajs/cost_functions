```python
%matplotlib notebook
import numpy as np
from glob import glob
from scipy.optimize import minimize, differential_evolution
from matplotlib import pyplot as plt

from costfun import *
```

```python
files = './data/*'
files = np.sort(glob(files))

vals = np.array([np.loadtxt(file) for file in files])


x = np.array([val[:,0] for val in vals])
y = np.array([val[:,1] for val in vals])

x, y = pad_vals(x,y)

sizelist = np.array([12, 14, 16, 18, 20])
```

```python
%%time
popsize0 = 100
maxiter0 = 1000
#x0 = np.zeros(len(sizelist) + 1, dtype=np.float64)
x0 = np.zeros(6, dtype = np.float64)
bounds = [(0., 5) for el in x0]
bounds[-1] = (0.1, 4)
# x0[0] = 0.54
# x0[1] = 0.10
# x0[-1] = 1.90
optRes = differential_evolution(minimization_fun,
                                bounds,
                                args=(x, y, sizelist, 'free', 'kt',
                                      False),
                                popsize=popsize0,
                                maxiter=maxiter0, workers=2)

print(optRes)
params = optRes.x
```
